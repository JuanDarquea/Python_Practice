```python
"""Lists practice"""
A = list()
count = 0
while True:
    n = input("Enter a number: ")
    if n == 'done':
        break

    try:
        nf = float(n)
        A.append(nf)
        count = count + 1
        continue
    except ValueError:
        print("The input must be a number")
    
average = sum(A)/count
print("List sum = ", sum(A))
print("List average = ", average)
print("List is:", A)
    
```

```python
"""Required file for the next assignment"""
from importlib.metadata import files
import os
import requests
from dotenv import load_dotenv

load_dotenv()

# Set up the file directory and URL
url = "https://www.py4e.com/code3/romeo.txt?PHPSESSID=0c4ea3c530e3ba4d25d71c7e88dcd738"
url2 = "https://www.py4e.com/code3/mbox-short.txt?PHPSESSID=d37feb65e5fac5c3349cc39a0a2e6ea4" 
filedir = os.getenv("coursera_mich_dir", ".")
data = requests.get(url).text
data2 = requests.get(url2).text

# Define the filename and path
filepath = os.path.join(filedir, "romeo.txt")
filepath2 = os.path.join(filedir, "mbox-short.txt")
files = [filepath, filepath2]

# Save the file if it does not exist
for file in files:
    if not os.path.exists(file):
        with open(file, 'w', encoding='utf-8') as f:
            if file == files[0]:
                f.write(data)
            else:
                f.write(data2)
    else:
        print(f"The file {file} already exists.")
```

```python
"""Create a sorted word list from a string"""
# Read the file
# Use the file name romeo.txt as the file name
fname = "romeo.txt"
fh = open(fname)
lst = list()
for line in fh:
    sline = line.strip().split()
    for word in sline:
        if word not in lst:
            lst.append(word)
lst.sort()
print(lst)
```

```python
"""Read thrpugh a list of emails, make a dictionary of counts of the number of times each email appears 
and print the email that appears the most often and its count"""
name = input("Enter file:")
if len(name) < 1:
    name = "mbox-short.txt"

try:
    handle = open(name)
except FileNotFoundError:
    print(f"File not found: {name}")
    raise

many = dict()

for line in handle:
    if not line.startswith('From:'):
        continue
    wds = line.split()
    if len(wds) < 2:
        continue
    email = wds[1]
    many[email] = many.get(email, 0) + 1

handle.close()

print(many)
# find the word with the largest count

bigword = None
largest = None
for key, val in many.items():
    if largest is None or val > largest:
        largest = val
        bigword = key

if bigword is not None:
    print(bigword, largest)
```
