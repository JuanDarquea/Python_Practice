```python
"""Lists practice"""
A = list()
count = 0
while True:
    n = input("Enter a number: ")
    if n == 'done':
        break

    try:
        nf = float(n)
        A.append(nf)
        count = count + 1
        continue
    except ValueError:
        print("The input must be a number")
    
average = sum(A)/count
print("List sum = ", sum(A))
print("List average = ", average)
print("List is:", A)
    
```

```python
"""Create a sorted word list from a string"""
# Read the file
# Use the file name romeo.txt as the file name
fname = "romeo.txt"
fh = open(fname)
lst = list()
for line in fh:
    sline = line.strip().split()
    for word in sline:
        if word not in lst:
            lst.append(word)
lst.sort()
print(lst)
```
