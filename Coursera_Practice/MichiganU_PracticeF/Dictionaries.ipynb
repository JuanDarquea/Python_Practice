```python
"""Dictionaries related utility functions and overall syntax."""
dicti1 = dict()
dicti1['key1'] = 6
dicti1['key2'] = 24
dicti1['key3'] = 12
print(dicti1)
dicti1['key2'] = dicti1['key2'] + 4
print(dicti1)
dicti1['key4'] = dicti1['key2'] + dicti1['key3']
dicti1['key1'] = 2
print(dicti1)
```

```python
"""Counting with dictionaries."""
dicti2 = dict()
names = ['Alice', 'Bob', 'Alice', 'Eve', 'Bob', 'Alice']
for name in names:
    if name not in dicti2:
        dicti2[name] = 1
    else:
        dicti2[name] = dicti2[name] + 1

print(dicti2)

"""Using the get() method for counting."""
dicti3 = dict()
for name in names:
    dicti3[name] = dicti3.get(name, 0) + 1
    
print(dicti3)
```

```python
"""Counting letters in a string using a dictionary and a input."""
string = input("Enter a string: ")
words = string.split()
words_count = dict()

for word in words:
    words_count[word] = words_count.get(word, 0) +1
print(words_count)

print("--"*20)

"""Printing the word counts in tuples sorted by word."""
for word, count in sorted(words_count.items()):
    print(f"{word}: {count}")
```

```python
"""Iterating through a dictionary."""
counts = {'key1': 3, 'key2': 5, 'key3': 2}
for key in counts:
    print(f"{key}: {counts[key]}")

print("--"*20)

"""Dictionary methods to retrieve keys, values, and items."""
print(list(counts)) # Returns list of keys
print(list(counts.keys())) # Returns list of keys
print(list(counts.values())) # Returns list of values
print(list(counts.items())) # Returns list of tuples

print("--"*20)

"""Printing the word counts in tuples sorted by word."""
for word, count in counts.items():
    print(f"{word}: {count}")
```
