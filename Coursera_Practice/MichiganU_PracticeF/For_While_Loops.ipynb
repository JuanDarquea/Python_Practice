```python
"""While loop example: countdown from 5 to Blastoff"""
n = 5
while n > 0:
    print(n)
    n = n - 1
print("Blastoff")
```

```python
"""While loop example: echo input until 'done' is entered"""
while True:
    line = input('> ')
    if line == 'done':
        break
    print(line)
print('Done')
```

```python
"""While loop example: echo input until 'done' is entered, ignoring comments"""
while True:
     # line = raw_input('> ')  # For Python 2 compatibility
     line = input('> ')
     if line[0] == '#':
         continue
     if line == 'done':
         break
     print(line)
print('Done')
```

```python
"""For loop example: countdown from 5 to Blastoff"""
for i in [5, 4, 3, 2, 1]:
    print(i)
print("Blastoff")
```

```python
"""For loop example: wish happy new year to friends"""
friends = ['Juan', 'David', 'Ana']
for a in friends:
    print("Happy new year:", a)
print('Done')
```

```python
"""For loop example: find the largest number in a list"""
largest_so_far = -1
print("Before", largest_so_far)
for num in [31, 3, 96, 1, 22, 10]:
    if num > largest_so_far:
        largest_so_far = num
    print(largest_so_far, num)
print("After", largest_so_far)
```

```python
"""While loop example: find the largest and smallest number from user input"""
largest = None
smallest = None
while True:
    num = input("Enter a number: ")
    if num == "done":
        break
    
    try:
        fnum = float(num)
    except:
        print("Invalid input")
        continue
        
    if largest is None or fnum > largest:
        largest = fnum
    if smallest is None or fnum < smallest:
        smallest = fnum

print("Maximum is", largest)
print("Minimum is", smallest)
```

```python
"""For loop example: iterate through a string and print each letter with its index"""
x = "Chronichle"
index = 0
for letter in x:
    print(letter, index)
    index = index + 1
```

```python
"""While loop example: iterate through a string and print each letter with its index"""
x = "Chronichle"
index = 0
while index < len(x):
    letter = x[index]
    print(letter, index)
    index = index + 1
```

```python
"""For loop example: iterate through a string and count occurrences of a letter"""
word = input("Enter the word to search: ")
let = input("Enter the letter to count: ")
count = 0
for letter in word:
    if letter == let or letter == let.upper():
        count = count + 1
print(f"The letter '{let}' appears {count} times in the word {word}")
```

```python
"""Slicing string example"""
x = input("Enter a word: ")
try:
    if type(x) != str:
        print("Input is not a string.")
    elif len(x) < 3:
        print("Word is less than 3 letters long.")
    else:
        print(f"Word is {x}")
        print("First 3 letters: ", x[0:3])
        print("Last 3 letters: ", x[-3:])
         
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
"""Slicing example: extract domain from email address"""
email = input("Enter your email address: ")
pieces = email.split('@')
domain = pieces[1]
print(f"The domain of the email address is: {domain}")
```

```python
"""Writing examples using string methods"""
greet = "Hello World"
print("E1:", greet.lower())
print("E2:", greet.upper())

greet2 = greet.replace("World", "Universe")
print("E3:", greet2)

greet3 = greet.lower()
print("E4:", greet3.capitalize())

line = "Please have a nice day"
print("E5:", line.startswith("Please"))
print("E6:", line.endswith("day"))
print("E7:", line.find("nice"))
print("E8:", line.startswith("p"))

line2 = greet.rstrip("World")
print("E9:", line2, f"/ {len(line2)} letters")
line3 = greet.lstrip("Hello")
print("E10:", line3, f"/ {len(line3)} letters")
```
